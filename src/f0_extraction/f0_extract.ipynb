```python
from scipy.io import wavfile
import pathlib
import audioflux as af
import numpy as np
import matplotlib.pyplot as plt
```

```python
target = list(pathlib.Path('../../assets/').glob('*.wav'))[0]
audio_arr, sr = af.read([target])
audio_arr.shape
audio_arr = audio_arr[0]
```

```python
pitch_obj = af.PitchPEF(samplate=sr)
fre = pitch_obj.pitch(audio_arr)
times = np.arange(fre.shape[-1]) * (pitch_obj.slide_length / sr)
fig, ax = plt.subplots(nrows=2, sharex=True)
af.display.fill_wave(audio_arr, samplate=sr, axes=ax[0])
ax[1].scatter(times, fre, s=2)
```
